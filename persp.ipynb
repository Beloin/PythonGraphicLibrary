```python
import raster
import wireframe
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

import transformation
```

```python
def extract_polygon(edges):
    poly = []
    verts = []
    counter = 0
    for vec3d in edges:
        p1 = vec3d[0]
        verts.append(p1)
        counter += 1

        if counter == edges.sep():
            poly.append(verts)
            verts = []
            counter = 0
    return poly
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

edges = wireframe.cube(.5, res=.5)

poly = []
verts = []
counter = 0
for vec3d in edges:
    p1 = vec3d[0]
    verts.append(p1)
    counter += 1

    if counter == 4:
        poly.append(verts)
        verts = []
        counter = 0

ax.add_collection3d(Poly3DCollection(poly, linewidths=1, alpha=0, edgecolors="black"))
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

edges = wireframe.cube(.5, res=.5)
edges = transformation.translate_edges(edges, (-.2, -.2, -.2))

poly = extract_polygon(edges)

ax.add_collection3d(Poly3DCollection(poly, linewidths=1, alpha=0, edgecolors="black"))
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

edges = wireframe.cube(.5, res=.5)
edges = transformation.rotate_edges(edges, (0, 20, 0))
edges = transformation.translate_edges(edges, (-.5, 0, 0))

poly = extract_polygon(edges)

ax.add_collection3d(Poly3DCollection(poly, linewidths=1, alpha=0, edgecolors="black"))
plt.show()
```
