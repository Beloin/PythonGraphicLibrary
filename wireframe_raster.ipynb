```python
import raster
import wireframe
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
# TODO: Yet we need to project on 2D plane
scale_x, scale_y = (500, 500)
scale = (scale_x, scale_y)

edge = wireframe.cube(.5, dist=.2)

polygon = raster.draw_edge_list(edge.raw(), scale)

mx = [[0 for _ in range(scale_x)] for _ in range(scale_y)]
raster.insert_points(mx, polygon)

plt.imshow(raster.convert_img(mx))
plt.show() 
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

edges = wireframe.cube(.5, dist=.2)
verts = []
for vec3d in edges:
    tmp = [0]
    tmp.extend(vec3d[0])
    verts.append(tmp)    

ax.add_collection3d(Poly3DCollection(verts))
plt.show()
```
