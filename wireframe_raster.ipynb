```python
import raster
import wireframe
import matplotlib.pyplot as plt
```

```python
# TODO: Yet we need to project on 2D plane
scale_x, scale_y = (500, 500)
scale = (scale_x, scale_y)

edge = wireframe.cube(.5, dist=.2)

for pt in edge:
    print(pt)

polygon = raster.draw_edge_list(edge.raw(), scale)

mx = [[0 for _ in range(scale_x)] for _ in range(scale_y)]
raster.insert_points(mx, polygon)

plt.imshow(raster.convert_img(mx))
plt.show() 
```
